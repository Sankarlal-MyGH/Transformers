```python
from transformers import pipeline
```

```python
classifier=pipeline('sentiment-analysis')
```

```python
classifier("We can watch the movie one time")
```

```python
results=classifier(["I am a good boy","I am a bad girl"])
results
```

```python
for result in results:
  print(result['label'],result['score'])
```

```python
classifier=pipeline('question-answering', model='distilbert-base-cased-distilled-squad')
```

```python
QA_input = {
    'question': 'Who will win IPL final?',
    'context': 'The option is either Chennai or Delhi.'
}
res=classifier(QA_input)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
  
tokenizer = AutoTokenizer.from_pretrained("google/pegasus-xsum")
model = AutoModelForSeq2SeqLM.from_pretrained("google/pegasus-xsum")
```

```python
text_summary_new=pipeline('summarization', model=model, tokenizer=tokenizer)
```

```python
from transformers import pipeline
```

```python
classifier=pipeline('sentiment-analysis')
```

```python
classifier("We can watch the movie one time")
```

```python
results=classifier(["I am a good boy","I am a bad girl"])
results
```

```python
for result in results:
  print(result['label'],result['score'])
```

```python
classifier_new=pipeline('sentiment-analysis', model='sampathkethineedi/industry-classification')
```

```python
classifier_new("Nazara tech profit is rised 30% today")
```
